```python
from catboost import CatBoostRegressor, CatBoostClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import DBSCAN, KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, minmax_scale
from sklearn.ensemble import StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
import seaborn as sns
```

# Читаем данные 

```python
X = pd.read_csv('train.csv', header=None)
y = pd.read_csv('train-target.csv', header=None)
test_orig = pd.read_csv('test.csv', header=None)
```

## Настраиваем валидацию

```python
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, shuffle=True, random_state=0)
```

## Смотрим на данные

```python
plt.hist(y) # распределение классов
```

```python
for i in range(30):
    plt.hist(X[i], color='orange', label=f'train[{i}]')
    plt.hist(test_orig[i], color='blue', label=f'test[{i}]')
    
    plt.xlabel('Значения')
    plt.ylabel('Частота')
    plt.title(f'{i} фича train и test')
    plt.legend()
    
    plt.show()
```

```python
df = pd.concat([X, y], axis=1)

corr = df.corr()
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
corr = test_orig.corr()
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

## Определяем наши фичи, на которых будем обучаться

```python
col2models = [i for i in range(30) if i not in [9, 11, 15, 16, 17, 22, 26]]
```

## BASELINE KNN

```python
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(X_train[col2models], y_train)
preds_knn = knn.predict_proba(X_test[col2models])
roc_auc_score(y_test, preds_knn[:, 1])
```

## Пробуем CatBoost

```python
cb = CatBoostClassifier(random_state=0, verbose=100)
cb.fit(X_train[col2models], y_train)
preds = cb.predict_proba(X_test[col2models])
roc_auc_score(y_test, preds[:, 1])
```

```python
cb = CatBoostClassifier(
        iterations=1500,
        max_depth=9,
        random_state=0,
        verbose=100
)
cb.fit(X_train[col2models], y_train)
preds_cb = cb.predict_proba(X_test[col2models])
roc_auc_score(y_test, preds[:, 1])
```

## MLP

```python
# Скейлим данные обязательно
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[col2models])
X_test_scaled = scaler.fit_transform(X_test[col2models])


mlp = MLPClassifier(hidden_layer_sizes=(16, 8), 
                    random_state=0, 
                   )
mlp.fit(X_train_scaled, y_train)
preds_mlp = mlp.predict_proba(X_test_scaled)
roc_auc_score(y_test, preds_mlp[:, 1])
```

```python
all_preds = pd.DataFrame()
all_preds['knn'] = preds_knn[:, 1]
all_preds['cb'] = preds_cb[:, 1]
all_preds['mlp'] = preds_mlp[:, 1]
all_preds
```

```python
corr = all_preds.corr()
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

## Модельки друг с другом не сильно коррелируют, пробуем ансамбли

```python
preds_cb_knn_mlp = (preds_cb + preds_knn + preds_mlp)/3
roc_auc_score(y_test, preds_cb_knn_mlp[:, 1])
```

```python
estimators = [
     ('cb1', CatBoostClassifier(
                iterations=1500,
                max_depth=9,
                feature_weights=cb.feature_importances_,
                random_state=0,
                verbose=100
        )),
    ('knn', knn),
    ('mlp', make_pipeline(StandardScaler(),
                           MLPClassifier(hidden_layer_sizes=(16, 8), 
                                         random_state=0, 
                                         solver='lbfgs',
                                         learning_rate='adaptive')))
]

clf = StackingClassifier(
         estimators=estimators, final_estimator=CatBoostClassifier(verbose=100, random_seed=0))
clf.fit(X_train[col2models], y_train)
preds_clf = clf.predict_proba(X_test[col2models])
roc_auc_score(y_test, preds_clf[:, 1])
```

## Обучаем на всех данных и отправляем сабмит!

```python
clf.fit(X[col2models], y)
preds = clf.predict_proba(test_orig[col2models])
pd.DataFrame(preds[:, 1]).to_csv('top4ik.csv', header=None, index=False)
```
